```python
import requests
import json
from pprint import pprint
import pandas as pd
```

```python
url = 'https://api.delivery-club.ru/api1.2/vendors/search?latitude=56.992744&longitude=40.973094&query=роллы&cacheBreaker=1663163124'
```

```python
response = requests.get(url)

print(response.status_code)
pprint(response.json())
response = response.json()['vendors']
```

```python
name_place = []
category_id = []
ave_min_price = []
price_delivery = []
for item in response:
    name_place.append(item['name'])
    category_id.append(item['categoryId'])
    ave_min_price.append(item['delivery']['minOrderPrice']['value'])
    price_delivery.append(item['delivery']['price']['value'])
print(name_place, category_id, ave_min_price, price_delivery)
```

```python
table = pd.DataFrame({'name_place': name_place,
                      'category_id': category_id,
                      'ave_min_price': ave_min_price,
                      'price_delivery': price_delivery})
table
```

```python
t1 = table[table['price_delivery'] == 0]
t1
```

```python
t2 = table[table['price_delivery'] != 0]
t2
```

```python
d1 = t1.shape[0]/table.shape[0]*100
print(f'Долю с бесплатной доставкой {d1}%')
```

```python
d2 = t2.shape[0]/table.shape[0]*100
print(f'Долю с платной доставкой {d2}%')
```

```python
p1 = t1['ave_min_price'].mean()
print(f'Средняя минимальная стоимость заказа с бесплатной доставкой {p1}')
```

```python
p2 = t2['ave_min_price'].mean()
print(f'Средняя минимальная стоимость заказа с платной доставкой {p2}')
```

```python
c1 = t1.groupby('category_id').agg({'category_id': 'count'})/t1.shape[0]
c1
```

```python
c2 = t2.groupby('category_id').agg({'category_id': 'count'})/t2.shape[0]
c2
```
