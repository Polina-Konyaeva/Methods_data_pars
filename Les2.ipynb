```python
from pprint import pprint
from lxml import html
import requests
```

```python
url = 'https://lenta.ru'


headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.167 YaBrowser/22.7.5.1027 Yowser/2.5 Safari/537.36',
}

response = requests.get(url, headers=headers)
pprint(response.text)
```

```python
dom = html.fromstring(response.text)
news_links =  dom.xpath("//div[@class='longgrid-list__box']/a[@class='card-mini _longgrid']/@href")
news_name = dom.xpath("//div[@class='longgrid-list__box']/a[@class='card-mini _longgrid']/div/span/text()")
news_time =  dom.xpath("//div[@class='longgrid-list__box']/a[@class='card-mini _longgrid']/div/div/time/text()")
```

```python
news ={}
for i in range(len(news_name)):
    news[i] = {
        'news_name':news_name[i],
        'news_links ':url+news_links[i] ,
        'news_time':news_time[i],
    }
```
